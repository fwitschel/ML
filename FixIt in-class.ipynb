```python
import numpy as np

# pandas allows to load data from csv files and handle them nicely
import pandas as pd

# classifier:
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# evaluation:
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import metrics

# visualisation of the decision tree
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus

# hack for python to better find GraphViz - might not be necessary on your system
# see https://datascience.stackexchange.com/questions/37428/graphviz-not-working-when-imported-inside-pydotplus-graphvizs-executables-not
import os
os.environ['PATH'] = os.environ['PATH']+';'+os.environ['CONDA_PREFIX']+r"\Library\bin\graphviz"
```

```python
fixit = pd.read_csv("fixit.csv")
fixit.head()
```

```python
feature_cols = ['Agent Experience','Agent Name','Customer Type','Priority','Problem Category','Ticket Type']
X = fixit[feature_cols]
y = fixit.SLAFullfillment
```

```python
one_hot_X = pd.get_dummies(X)
one_hot_X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(one_hot_X,y,test_size=0.33)
dtree = DecisionTreeClassifier(max_depth=4)

# train Decision Tree classifer
dtree = dtree.fit(X_train,y_train)

# make the predictions on the test data
y_pred_tree = dtree.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_tree))
```

```python
# finally, let's visualise this using GraphViz:
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = list(X_train.columns),class_names=['IN_RANGE','SLA_VIOLATION'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
